```python
import qiskit
import numpy as np
from qiskit import Aer, QuantumCircuit, execute, ClassicalRegister, QuantumRegister, BasicAer
from qiskit.visualization import plot_bloch_multivector, plot_histogram
from math import pi, sqrt
```

```python
qiskit.__qiskit_version__
```

```python
Aer.backends()
```

```python
BasicAer.backends()
```

```python
cr = ClassicalRegister(1,"cr")
qr = QuantumRegister(1,"qr")
qc = QuantumCircuit(qr,cr)
```

```python
qc.draw()
```

```python
qc.x(0)
```

```python
qc.draw()
```

```python
backend = Aer.get_backend('statevector_simulator')
result = execute(qc,backend).result().get_statevector(qc,decimals=4)
print(result)
```

```python
plot_bloch_multivector(result)
```

```python
backend = Aer.get_backend('unitary_simulator')
result = execute(qc,backend).result().get_unitary(qc,decimals=4)
print(result)
```

```python
plot_bloch_multivector(result)
```
